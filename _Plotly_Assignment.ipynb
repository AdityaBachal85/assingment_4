```python
import seaborn as sns
import plotly.express as px

titanic = sns.load_dataset("titanic")
fig = px.scatter(titanic, x="age", y="fare")
fig.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go

tips = px.data.tips()
fig = px.box(tips, y="total_bill")
fig.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go

tips = px.data.tips()
fig = px.histogram(tips, x="sex", y="total_bill", color="smoker", pattern_shape="smoker", facet_col="day", color_discrete_map={"Yes": "blue", "No": "red"})
fig.show()
```

```python
import plotly.express as px

iris = px.data.iris()
fig = px.scatter_matrix(iris, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"], color="species")
fig.show()
```

```python
import plotly.express as px

tips = px.data.tips()
fig = px.histogram(tips, x="total_bill", marginal="rug", histnorm="probability density")
fig.show()
```
